Stáhni si soubor platy_2021_02.csv s informacemi o platech v softwarové firmě (stejný soubor, jako v předchozím úkolu).
Načti si tato data do tabulky a vytvoř histogram. Nastav vhodně hranice skupin histogramu (parametr bins), aby byl graf přehledný a snadno interpretovatelný.

```python
platy = pd.read_csv("platy_2021_02.txt",sep=",")
```

Zjištění nejvyššího a nejnižšího platu

```python
extremy = [platy.min(),platy.max()]
extremy
```

Histogram

```python
plt = platy["plat"].plot(kind="hist", range=[30_000, 60_000], bins=6, color="royalblue", edgecolor = "black", title = "Rozložení platů ve firmě")

plt.set_xlabel("Plat (Kč)")
plt.set_ylabel("Četnost")
```
