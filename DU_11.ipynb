## Povinná část

```python
import pandas as pd
```

Stáhni si soubor platy_2021_02.csv s informacemi o platech v softwarové firmě (stejný soubor, jako v předchozím úkolu).
Načti si tato data do tabulky a vytvoř histogram. Nastav vhodně hranice skupin histogramu (parametr bins), aby byl graf přehledný a snadno interpretovatelný.

```python
platy = pd.read_csv("platy_2021_02.txt",sep=",")
```

Zjištění nejvyššího a nejnižšího platu

```python
extremy = [platy.min(),platy.max()]
extremy
```

Histogram

```python
plt = platy["plat"].plot(kind="hist", range=[30_000, 60_000], bins=6, color="royalblue", edgecolor = "black", title = "Rozložení mezd ve firmě")

plt.set_xlabel("Plat (Kč)")
plt.set_ylabel("Počet zaměstnanců")
```

## Nepovinný bonus 1

Vrať se k práci se souborem temperature.csv, který obsahuje informace o průměrné teplotě v různých městech v listopadu 2017.

```python
teploty = pd.read_csv("temperature.txt",sep=",")
teploty.head()
```

Vytvoř tabulku, která bude obsahovat údaje o teplotě za města Helsinki, Miami Beach a Tokyo.

```python
vybrana_mesta = teploty[teploty["City"].isin(["Helsinki","Miami Beach","Tokyo"])]
vybrana_mesta.head()
```

Vytvoř krabicový graf a porovnej rozsah teplot v těchto městech.

```python
bplt = vybrana_mesta.boxplot(by ="City", column =["AvgTemperature"], grid = False)
#bplt.suptitle(" ")
bplt.set_title("Průměrné teploty ve vybraných městech")
bplt.set_xlabel(" ")
bplt.set_ylabel("Průměrná teplota (\N{DEGREE SIGN}F)")
```
